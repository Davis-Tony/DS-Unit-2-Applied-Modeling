```python
#Reading in data
import pandas as pd

url='https://raw.githubusercontent.com/Davis-Tony/DS-Unit-2-Applied-Modeling/master/WhoopData.csv'

df = pd.read_csv(url)
```

```python
df.head()
```

```python
#Beginning data explortaion
#Checking for missing data
df.isnull().sum()
```

```python
#dropping nap columns as they're nearly totally empty
df = df.drop(['whoop_nap_count', 'whoop_nap_duration_total_asleep'], axis=1)
```

```python
df.head()
```

```python
#Dropping first row as most of the info included is NaN
df = df.drop([0])
```

```python
df.head()
```

```python
df.dtypes
```

```python
#Converting date to datetime format
df['date'] = pd.to_datetime(df['date'])
```

```python
df.dtypes
```

```python
df.head()
```

```python
#Creating new feature to qualify great sleep
df['great_recovery'] = df['whoop_recovery_score'] > 66
```

```python
#Verifying
df.head()
```

```python
df['great_recovery'].value_counts(normalize=True)
```

This will be a binary classification problem as the target will be the 'great_recovery' column split in two classes. With the majority class at 62%, this is a pretty balanced split.

Since it's balanced, we'll use accuracy as the evaluation metric.

I'll be using a time-based split using a specific date as the split between test and train datasets.

The one feature that would result in leakage is the 'whoop_recovery_score' column beacuse that directly feeds the targeted 'great_recovery' column.
